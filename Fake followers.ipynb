```python
import numpy as np
import pandas as pd
import itertools
```

```python
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')
```

```python
train_data
```

```python
test_data
```

```python
x_train = train_data.drop(columns='fake')
x_train
```

```python
y_train = train_data.iloc[:,-1]
y_train
```

```python
x_test = test_data.drop(columns='fake')
x_test
```

```python
y_test = test_data['fake']
```

```python
y_test
```

```python
x_data ,y_data = x_train,y_train
```

```python
x_data
```

```python
y_data
```

```python
x_data.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_data,y_data)
```

```python
x_train
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

###  Support Vector Machine

```python
from sklearn.svm import SVC
Svc = SVC()
Svc.fit(x_train,y_train)
y_pred = Svc.predict(x_test)
svc_score = accuracy_score(y_pred,y_test)
print(f'Accuracy= {round(svc_score*100,2)}%')
```

### Logisitic Regression

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(x_train,y_train)
log_pred = log_reg.predict(x_test)
log_score = accuracy_score(log_pred,y_test)
print(f'Accuracy = {round(log_score*100,2)}%')
```

### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rf_clf = RandomForestClassifier()
rf_clf.fit(x_train,y_train)
rf_pred = rf_clf.predict(x_test)
rf_score = accuracy_score(rf_pred,y_test)
print(f'Accuracy = {round(rf_score*100,2)}%')
```

### Gradient Boosting Classifier

```python
from sklearn.ensemble import GradientBoostingClassifier
gb_clf = GradientBoostingClassifier()
gb_clf.fit(x_train,y_train)
gb_pred = gb_clf.predict(x_test)
gb_score = accuracy_score(gb_pred,y_test)
print(f'Accuracy= {round(gb_score*100,2)}%')
```

### Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x_train,y_train)
nb_pred = nb.predict(x_test)
nb_score = accuracy_score(nb_pred,y_test)
print(f'Accuracy= {round(nb_score*100,2)}%')
```

### Average Accuracy

```python
avg_acc = (svc_score+log_score+rf_score+gb_score+nb_score)/5
print(f'Average Accuracy = {round(avg_acc*100,2)}%')
```

### Confusion Matrix

```python
cm = confusion_matrix(y_test,nb_pred)
print(cm)
```
